### sorted() in python
- sorted function returns the sorted list of an iterable object

# sorted(iterable, key, reverse)
- 'iterable' is required parameter. This is iterable object to sort(list, tuple, dict)
- key is optional. Default value is none, this key is to decide the execution order.
- revers is optional. Default value is false, to sort based on ascending or descending order. Default is ascending order.

```python
# Examples demonstrating sorted() with the 'key' parameter
# 1) Sort list of tuples by the second element
data = [(1, 'b'), (3, 'a'), (2, 'c')]
print('Original tuples:', data)
print('Sorted by second element:', sorted(data, key=lambda x: x[1]))

# 2) Case-insensitive sorting of strings using key=str.lower
names = ['alice', 'Bob', 'charlie', 'bob']
print('Case-sensitive sort:', sorted(names))
print('Case-insensitive sort:', sorted(names, key=str.lower))

# 3) Sort list of dictionaries by a specific dictionary key
people = [{'name': 'Alice', 'age': 30}, {'name': 'Bob', 'age': 25}, {'name': 'Charlie', 'age': 35}]
print('Original people:', people)
print('Sorted by age:', sorted(people, key=lambda p: p['age']))

# 4) Sort by absolute value (example with numbers) and reverse order
nums = [-10, 3, -2, 7, 0]
print('Sorted by absolute value:', sorted(nums, key=abs))
print('Sorted by absolute value (descending):', sorted(nums, key=abs, reverse=True))
```

# Lambda Functions
* functions with no name, kind of one line anonymous function
# lambda variables : operation on variables
* lambda x : x+y

```python
# Lambda examples
# 1) simple lambda for addition
add = lambda x, y: x + y
print('add(2, 3) =', add(2, 3))

# 2) use lambda with map to square numbers
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x * x, nums))
print('squares:', squares)

# 3) filter even numbers using lambda
evens = list(filter(lambda x: x % 2 == 0, nums))
print('evens:', evens)

# 4) reduce to compute product (requires functools.reduce)
from functools import reduce
product = reduce(lambda a, b: a * b, nums)
print('product:', product)

# 5) sort with lambda key (sort by last character)
words = ['apple', 'banana', 'cherry', 'date']
print('sorted by last char:', sorted(words, key=lambda w: w[-1]))
```

## Example for positional argument

```python
def greet(name, message):
    print(f'Hello {name}, {message}')

# Using positional arguments
greet('Alice', 'welcome to the session!')
greet('Bob', 'good morning!')
```

## Example for keyword argument

```python
def greet(name, message):
    print(f'Hello {name}, {message}')

# Using keyword arguments
greet(name='Alice', message='welcome to the session!')
greet(message='good morning!', name='Bob')
```
