# Loops In Python

```python
# if Loop
while True:
    s = input("Enter a number: ").strip()
    if s == '':
        print("No input provided. Please enter an integer.")
        continue
    try:
        num = int(s)
        break
    except ValueError:
        print(f"Invalid integer: {s!r}. Please try again.")
if num > 0:
    print(f"{num} is positive")
else:
    print(f"{num} is zero or negative")

# if-elif-else Loop
while True:
    st = input("Enter a score (0-100): ").strip()
    if st == '':
        print("No input provided. Please enter an integer.")
        continue
    try:
        score = int(st)
        if 0 <= score <= 100:
            break
        else:
            print("Score must be between 0 and 100. Please try again.")
    except ValueError:
        print(f"Invalid integer: {st!r}. Please try again.")
if score >= 90:
    grade = 'A'
elif score >= 75:
    grade = 'B'
elif score >= 60:
    grade = 'C'
else:
    grade = 'F'

print(f"Score: {score} -> Grade: {grade}")
```

# Range Function with for loop
 -  range(start, stop, step) -> gives values within two numbers

```python
rangeValue = range(10)
print("Range Value for provided input ",rangeValue)

for i in range(1,10):
    print(i)

for i in range(100, -1, -15):
    print(i)
```

```python
# Program for writing mulitiplication table of a given number
num = int(input("Enter a number to print its multiplication table: "))
for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

# Build In Data structures
 - List
 - Tuple
 - Set
 - Dictionary

```python
# List of Built In Data structures

balanceList = [1000, 2500, 3000, 4000, 5000]
updatedbalanceList = [1001, 2501, 3001, 4001]
print("Balance List: ", balanceList)
print("Type of balanceList is: ", type(balanceList))
print("Length of balanceList is: ", len(balanceList))
print("First element in balanceList is: ", balanceList[0])
print("Last element in balanceList is: ", balanceList[-1])

#iterable in list
for balance in balanceList:
    print("Iterable output: ", balance)

# To add element in list
balanceList.append(6000)
print("After append balanceList: ", balanceList)
balanceList.insert(2, 2750)
print("After insert balanceList: ", balanceList)

# To extend another list to parent list
balanceList.extend(updatedbalanceList)
print("After extend balanceList: ", balanceList)

# Range Function in List
# use a different loop variable name so we don't overwrite the list variable
for bal in range(balanceList[0], balanceList[-1] + 1, 1000):
    print("Range function output: ", bal)
```
