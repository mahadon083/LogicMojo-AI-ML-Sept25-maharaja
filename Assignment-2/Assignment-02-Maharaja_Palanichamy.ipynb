## Problem: 1
Create a uniform subdivision of the interval -1.3 to 2.5 with 64 subdivisions.

```python
def findUniformSubdivisions(start, end, subdivisions_count):
    subdivisions = np.linspace(start, end, subdivisions_count)
    return subdivisions
print(findUniformSubdivisions(-1.3, 2.5, 64))
```

## Problem: 4

Create a 10 x 10 arrays of zeros and then "frame" it with a border of ones.

```python
def createBorderedArray():
    array = np.zeros((10, 10), dtype=int)
    array[0, :] = 1
    array[-1, :] = 1
    array[:, 0] = 1
    array[:, -1] = 1
    print(array)

createBorderedArray()
```

## Problem: 5

Create an 8 x 8 array with a checkerboard pattern of zeros and ones.

```python
# Create an 8 x 8 array with a checkerboard pattern of zeros and ones
def createCheckerboardArray():
    checkerboard = np.zeros((8, 8), dtype=int)
    checkerboard[1::2, ::2] = 1
    checkerboard[::2, 1::2] = 1
    print(checkerboard)

createCheckerboardArray()

```

# Problem: 8
Evaluate $cos$ and $sin$ on the interval $[0,1]$  with 1000 points and then stack the results into a tall array with rows being the $(cos(x),sin(x))$ entries.

```python
# Evaluate cos and sin on the interval [0,1] with 1000 points and then stack the results into a tall array with rows being the (cos(x),sin(x)) entries

def evaluateTrigFunctions(start, end, num_points):
    x = np.linspace(start, end, num_points)
    cos_values = np.cos(x)
    sin_values = np.sin(x)
    stacked_array = np.column_stack((cos_values, sin_values))
    print(stacked_array)

evaluateTrigFunctions(0, 1, 1000)
```

# Problem: 9

Create a random $3×5$ array using the `np.random.rand(3 ,5)` function and compute: the sum of all the entries, the sum of the rows, and the sum of the columns.

```python
def computeRandomArraySums(rows, cols):
    random_array = np.random.rand(rows, cols)
    total_sum = np.sum(random_array)
    row_sums = np.sum(random_array, axis=1)
    col_sums = np.sum(random_array, axis=0)
    print("Random Array:\n", random_array)
    print("Total Sum:", total_sum)
    print("Row Sums:", row_sums)
    print("Column Sums:", col_sums)

computeRandomArraySums(3, 5)
```

## Problem: 10

Create a random $5×5$ array using the function `np.random.rand(5, 5)`. Then sort the rows according to the second column.

```python
def sortArrayBySecondColumn(rows, cols):
    random_array = np.random.rand(rows, cols)
    print("Original Array:\n", random_array)
    sorted_array = random_array[random_array[:, 1].argsort()]
    print("Sorted Array by Second Column:\n", sorted_array)

sortArrayBySecondColumn(5, 5)
```
