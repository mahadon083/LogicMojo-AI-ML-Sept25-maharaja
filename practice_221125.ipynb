# Data types in Python
 - integer(int)
 - floating number(float)
 - String(str)
 - Boolean(bool)

# Naming Convention Rules for Data Types
  * No space
  * No special characters except _
  * underscore can be used anywhere in data type
  * Alphanumeric
  * Numbers cannot be in start position of variable
  * No reserved keyword can be used as data type name

```python
a=8.5
b=8
c="Name"
print(a)
print(b)
print(c)
```

Arithmetic Operations - Addition and Subtraction

```python
a1 = 8
a2 = 100.5
b1 = 200.3
b2 = 80

c1=a1+a2
print("Addition of a1 and a1 - ",c1)

c2=b2-b1
print("Subtraction of b1 and b2 - ",c2)
```

Other Arithmetic Operations

```python
a1 = 27
a2 = 100.5
b1 = 200.3
b2 = 83

# Multiplication
c3=a1*b2
print("Multiplication of a1 and b2 - ",c3)

# Division
c4=b1/a2
print("Division of b1 and a2 - ",c4)

# Module Division
c5=b1%b2
print("Module Division of b1 and b2 - ",c5)

# Exponent
c6=a2**2
print("Exponent of a2 - ",c6)

#round function
c7=round(b1)
print("Round of b1 - ",c7)
```

# Exploring Colorama library

```python
from colorama import Fore, Style  # Import Colorama

b1 = 200.3
b2 = 80

c2 = b2 - b1
print("Subtraction of b1 and b2 - ", c2)
if c2 >= 0:
    print(Fore.GREEN + "Subtraction result is Positive: ", c2, Style.RESET_ALL)
else:
    print(Fore.RED + "Subtraction result is Negative: ", c2, Style.RESET_ALL)
```
